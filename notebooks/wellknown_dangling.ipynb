```python
import os
import json
import requests
import tldextract
from tqdm import tqdm

for t in os.listdir("../data/wellknown/webauthn/"):

    with open(f"../data/wellknown/webauthn/{t}") as f:
        webauthn = json.load(f)

    print(f"# Checking {t}")
    for e in webauthn:
        origin = e["origin"]
        origins = e["data"]["origins"]

        origin_fqdn = tldextract.extract(origin).fqdn
        origins_fqdn = [tldextract.extract(o).fqdn for o in origins]

        # print(f"Origin: {origin_fqdn}")
        # print(f"Origins: {origins_fqdn}")

        for o in [origin_fqdn, *origins_fqdn]:
            # print(f"Checking: {o}")
            if not o:
                continue
            r = requests.get(f"https://dns.google/resolve?name={o}")
            data = r.json()
            if not "Answer" in data:
                print(f"\tDead domain discovery for: {o}")
                continue
            if len(data["Answer"]) == 0:
                print(f"\tDead domain discovery for: {o}")
                continue

    print("\n\n")
```
