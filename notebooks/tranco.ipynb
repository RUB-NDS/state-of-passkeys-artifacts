```python
import tranco

# Initialize Tranco list and get the latest ranking
t = tranco.Tranco(cache=True, cache_dir=".tranco-cache")
latest_list = t.list()

# Read sites from file
with open("sites.txt", "r") as f:
    lines = f.readlines()

# Process each line
results = []
for line in lines:
    line = line.strip()
    if not line:
        continue

    parts = line.split("\t")
    if len(parts) >= 2:
        name = parts[0]
        domain = parts[1]

        # Skip entries with N/A domain
        if domain == "N/A":
            tranco_rank = "N/A"
        else:
            # Get Tranco rank
            rank = latest_list.rank(domain)
            tranco_rank = str(rank) if rank != -1 else "N/A"

        # results.append(f"{name} {domain} {tranco_rank}")
        results.append(f"{tranco_rank}")

# Print all results
for result in results:
    print(result)
```
